# **Handling Outliers**

```python
q1 = df_1["rate"].quantile(0.25)
q3 = df_1["rate"].quantile(0.75)

iqr = q3 - q1

lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr

outliers = df_1[(df_1["rate"] < lower) | (df_1["rate"] > upper)]
```
